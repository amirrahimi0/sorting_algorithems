<p dir="rtl" style="text-align: justify;">   
    <b>Merge Sort Algorithm</b>
</p>
</font>
Merge Sort یکی از محبوب‌ترین الگوریتم‌های مرتب‌سازی است که بر اساس اصل الگوریتم Divide and Conquer بنا شده است. الگوریتم‌های بازگشتی اغلب از روش Divide and Conquer استفاده می‌کنند. در این روش، مسئله به زیرمسائل کوچک‌تر تقسیم می‌شود، زیرمسائل را به صورت بازگشتی حل می‌کنیم و سپس حل‌ها را برای ایجاد راه‌حل نهایی ترکیب می‌کنیم. اگر مسئله به اندازه کافی کوچک باشد، مستقیماً آن را حل می‌کنیم. در غیر این صورت، سه مرحله بازگشتی را انجام می‌دهیم:
<ol>
  <li ><b>Divide</b> : مسئله را به یک یا چند زیر مسئله تقسیم می کنیم که نمونه های کوچکتری از همان مسئله هستند.</li>
  <li><b>Conquer</b> : زیرمسائل را با حل بازگشتی آنها حل می کنیم.</li>
  <li><b>Combine</b> : راه حل های زیرمسئله را با هم ترکیب می کنیم تا راه حلی برای مسئله اصلی بدست آید.</li>
</ol>
الگوریتم Merge Sort به دقت از روش Divide and Conquer پیروی می کند. در هر مرحله، یک زیرآرایه را مرتب می کند، که از کل آرایه  شروع می شود و به زیر آرایه های کوچکتر و کوچکتر می رسد.
<hr>
</div>

$$
\begin{array}{|c|c|c|c|}
\hline \text { Worst Time } & \text { Best Time } & \text { Average } & \text { Sorting Algorithm } \\
\hline O(n \log n) & O(n \log n) & O(n \log n) & \text { Merge Sort } \\
\hline O\left(n^2\right) & O(n \log n) & O(n \log n) & \text { Quick Sort } \\
\hline O\left(n^2\right) & O(n) & O\left(n^2\right) & \text { Bubble Sort } \\
\hline O\left(n^2\right) & O(n) & O\left(n^2\right) & \text { Insertion Sort } \\
\hline
\end{array}
$$

</p>

<font size=5>
<p dir="rtl" style="text-align: justify;">
<b>
چگونگی عملکرد Merge sort
</b>
</p>
</font>
<p dir="rtl" style="text-align: justify;">
استفاده از تکنیک  Divide and Conquer، به ما این امکان را می‌دهد که یک مسئله را به زیرمسائل تقسیم کنیم. وقتی که حل هر زیرمسئله انجام شد، ما نتایج زیرمسائل را "ترکیب" کرده و با استفاده از آن‌ها مسئله اصلی را حل می‌کنیم.
</p>
<p dir="rtl" style="text-align: justify;">
فرض کنید که باید یک آرایه A را باید مرتب کنیم. یک زیر مسئله مرتب‌سازی یک بخش زیری از این آرایه است که از اندیس p شروع شده و در اندیس r پایان می‌یابد، که با A[p: r] نشان داده می‌شود.
</br>
</p>
<p dir="rtl" style="text-align: justify;">
<b>1.Divide</b> : اگر q نقطهٔ نصفه‌ آرایه باشد، می‌توانیم آرایه A[p: r] را به دو زیر آرایه A[q+1: r]  و A[p: q] تقسیم کنیم.
</p>
<p dir="rtl" style="text-align: justify;">
<b>2.Conquer</b> : در مرحله conquer سعی می کنیم هر دو زیر آرایه بالا را مرتب کنیم. اگر هنوز به حالت پایه نرسیده ایم، دوباره هر دو زیر آرایه را تقسیم می کنیم و سعی می کنیم آنها را مرتب کنیم.
</p>
<p dir="rtl" style="text-align: justify;">
<b>3.Combine</b> : هنگامی که مرحله conquer به مرحله پایه رسید و دو زیر آرایه مرتب شده به دست می‌آید، نتایج را با ایجاد یک آرایه مرتب‌شده A ترکیب می‌کنیم. [p..r] از دو زیر آرایه مرتب شده .
</p>
<p dir="rtl" style="text-align: justify;">
Merge sort به صورت بازگشتی آرایه را به دو نیم تقسیم می کند تا زمانی که به حالت پایه آرایه با 1 عنصر برسیم. پس از آن، تابع merge زیر آرایه های مرتب شده را انتخاب می کند و آنها را ادغام می کند تا به آرایه مرتب برسد.
</p>

<b>
پیاده سازی الگوریتم <font color = #D93137> Merge Sort </font>
</b>

```python
def mergeSort(arr):
	if len(arr) > 1:
		
		q = len(arr)//2
		L = arr[:q]
		R = arr[q:]

		mergeSort(L)
		mergeSort(R)

		i = j = k = 0
		while i < len(L) and j < len(R):
			if L[i] <= R[j]:
				arr[k] = L[i]
				i += 1
			else:
				arr[k] = R[j]
				j += 1
			k += 1

		while i < len(L):
			arr[k] = L[i]
			i += 1
			k += 1

		while j < len(R):
			arr[k] = R[j]
			j += 1
			k += 1
```

<ul dir="rtl" style="text-align: justify; ">
  <li>چک می‌کند که طول آرایه بیشتر از یک باشد، اگر آرایه تنها یک عضو دارد به این معناست که مرتب است و نیازی به مرتب‌سازی ندارد.</li>
  <li>تقسیم طول آرایه بر دو و نتیجه را به q اختصاص می‌دهد. این کار برای تقسیم آرایه به دو نیمه است.</li>
  <li>تعیین نیمه اول آرایه به عنوان L.</li>
  <li>تعیین نیمه دوم آرایه به عنوان R.</li>
  <li>فراخوانی بازگشتی تابع mergeSort بر روی نیمه اول آرایه.</li>
  <li>فراخوانی بازگشتی تابع mergeSort بر روی نیمه دوم آرایه.</li>
</ul>






```python
arr = [6, 12, 66, 44, 5, 77]

if len(arr) > 1:	
	q = len(arr)//2
	L = arr[:q]
	R = arr[q:]
```

<b>
قسمت <font color="#D93137">Conquer</font> کد
</b>
</p>

<ul dir="rtl" style="text-align: justify; ">
  <li>تعیین شمارنده‌های مورد نیاز برای حرکت در آرایه‌های L، R و arr.</li>
  <li>حلقه اجرایی که تا زمانی که هر دو نیمه هنوز اعضای باقی‌مانده دارند ادامه می‌یابد.</li>
  <li>اگر عنصر i ام از نیمه اول کوچکتر مساوی عنصر j ام از نیمه دوم باشد.</li>
  <li>عنصر i ام از نیمه اول به عنوان کوچکترین عنصر در arr در موقعیت k ام قرار می‌گیرد.</li>
  <li>شمارنده نیمه اول را افزایش می‌دهیم.</li>
  <li>اگر عنصر j ام از نیمه دوم کوچکتر باشد.</li>
  <li>عنصر j ام از نیمه دوم به عنوان کوچکترین عنصر در arr در موقعیت k ام قرار می‌گیرد.</li>
  <li>شمارنده نیمه دوم را افزایش می‌دهیم.</li>
  <li>شمارنده arr را افزایش می‌دهیم.</li>
  <li>حلقه اجرایی برای عناصر باقی‌مانده در نیمه اول.</li>
  <li> حلقه اجرایی برای عناصر باقی‌مانده در نیمه دوم.</li>
</ul>






```python
i = j = k = 0
while i < len(L) and j < len(R):
	
	if L[i] <= R[j]:
		arr[k] = L[i]
		i += 1
	else:
		arr[k] = R[j]
		j += 1
	k += 1

while i < len(L):
	arr[k] = L[i]
	i += 1
	k += 1

while j < len(R):
	arr[k] = R[j]
	j += 1
	k += 1
```
